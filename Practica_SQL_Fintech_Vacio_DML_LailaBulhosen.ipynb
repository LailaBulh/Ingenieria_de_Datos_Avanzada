## **Ejercicio 3. Transacciones**


- Obt√©n todas las transacciones de la cuenta **A1001**.  
- Calcula el **saldo neto** de esa cuenta sumando la columna `monto`.  
- Lista las transacciones que sean **comisiones**.  


````python
### Transacciones de la cuednta A1001
%%sql SELECT cuenta_id ,COUNT(tx_id) AS transactions FROM transacciones WHERE cuenta_id = 'A1001';
````

````python
### Saldo neto de cuenta A1001
%%sql SELECT cuenta_id ,SUM(monto) AS saldo_neto FROM transacciones WHERE cuenta_id = 'A1001'
````

````python
###Transacciones tipo Comisi√≥n
%%sql SELECT * FROM transacciones WHERE categoria = 'Comisi√≥n'
````

## **Ejercicio 4. Reportes con agrupaciones**


- Calcula el total de transacciones (suma de `monto`) por **categor√≠a**.     


````python
### Total de monto por categoria
%%sql SELECT categoria ,SUM(monto) AS monto FROM transacciones GROUP BY categoria;
````
