### **Ejercicio 1**

**CountVectorizer**

```python
### Convertir a minusculas y quitar acentos
vectorizer1 = CountVectorizer(lowercase = False, strip_accents = None)

X = vectorizer1.fit_transform(corpus_es)
```

```python
print(X)
```

```python
### Convertir arreglo a dataframe
bow_df = pd.DataFrame(X.toarray(), columns = vectorizer1.get_feature_names_out())

bow_df_matriz = bow_df.to_numpy()

display(bow_df)
```

### **Ejercicio 2**

```python
### Funcion para convertir minusculas, quitar stop words y quitar acentos

def remove_stopwords(text):
  text = text.lower()
  stopwords_es = stopwords.words('spanish')
  tokens = word_tokenize(text, language = 'spanish')
  clean_text = [word for word in tokens if word not in stopwords_es and word.isalpha()]
  clean_text_all = ' '.join(clean_text)
  return unidecode(clean_text_all)
```

```python
### Convertir a minusculas y quitar acentos
vectorizer_min = CountVectorizer(lowercase = True, preprocessor = remove_stopwords)

X_min = vectorizer_min.fit_transform(corpus_es)
```

```python
print(X_min)
```

```python
### Convertir a dataframe
corpus_df = pd.DataFrame(X_min.toarray(), columns = vectorizer_min.get_feature_names_out())

display(corpus_df)
```

```python
print(f'Dimensiones de la matriz de CountVectorizer: {corpus_df.shape}')
```

```python
all_tokens3 = vectorizer_min.get_feature_names_out()

for i, t in enumerate(all_tokens3):
  if i >= 10:
    break
  print(f'Token {i}: {t}')
```

**TfidVectorizer**

```python
### Implementar TF-IDF
vectorizer_tfidf = TfidfVectorizer(lowercase = True, strip_accents = 'unicode',preprocessor = remove_stopwords)

X2_tfidf = vectorizer_tfidf.fit_transform(corpus_es)
```

```python
### Convertir a dataframe
tfid_df2 = pd.DataFrame(X2_tfidf.toarray(), columns = vectorizer_tfidf.get_feature_names_out())

tfid_df_matriz = tfid_df2.to_numpy()

display(tfid_df2)
```

```python
print(f'Dimensiones de la matriz de TfidfVectorizer: {tfid_df2.shape}')
```

```python
all_tokens4 = vectorizer_tfidf.get_feature_names_out()

for i, t in enumerate(all_tokens4):
  if i >= 10:
    break
  print(f'Token {i}: {t}')
```
